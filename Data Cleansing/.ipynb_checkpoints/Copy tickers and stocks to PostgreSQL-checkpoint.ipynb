```python
# --- Copy tickers and stocks tables from SQLite to PostgreSQL ---

print("\n" + "="*80)
print("COPY TICKERS AND STOCKS FROM SQLITE TO POSTGRESQL")
print("="*80)

# Function to check if table exists in PostgreSQL
def table_exists(connection, table_name):
    try:
        sql = f"""
        SELECT EXISTS (
            SELECT FROM information_schema.tables 
            WHERE table_name = '{table_name.lower()}'
        )
        """
        result = pd.read_sql(sql, connection)
        return result.iloc[0, 0]
    except Exception as e:
        print(f"Error checking table {table_name}: {e}")
        return False

# Function to get table row count
def get_row_count(connection, table_name):
    try:
        sql = f"SELECT COUNT(*) FROM {table_name}"
        result = pd.read_sql(sql, connection)
        return result.iloc[0, 0]
    except Exception as e:
        print(f"Error getting row count for {table_name}: {e}")
        return 0

# Function to drop table with CASCADE
def drop_table_cascade(connection, table_name):
    try:
        connection.execute(text(f"DROP TABLE IF EXISTS {table_name} CASCADE"))
        connection.commit()
        print(f"  - Dropped table {table_name} with CASCADE")
        return True
    except Exception as e:
        print(f"  - Error dropping {table_name}: {e}")
        return False

# Function to truncate table with CASCADE
def truncate_table_cascade(connection, table_name):
    try:
        connection.execute(text(f"TRUNCATE TABLE {table_name} RESTART IDENTITY CASCADE"))
        connection.commit()
        print(f"  - Truncated table: {table_name}")
        return True
    except Exception as e:
        print(f"  - Error truncating {table_name}: {e}")
        return False

# Rollback any pending transactions to get a clean state
print("\nCleaning up PostgreSQL connection state...")
try:
    conpg.rollback()
    print("  - Rolled back any existing transaction")
except Exception as e:
    print(f"  - Note: {e}")

# 1. Create backup of existing PostgreSQL tables
print("\n" + "-"*60)
print("1. CREATING BACKUPS OF EXISTING POSTGRESQL TABLES")
print("-"*60)

backup_folder = dat_path
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# Backup existing tickers if it exists
if table_exists(conpg, 'tickers'):
    df_tickers_pg_backup = pd.read_sql("SELECT * FROM tickers ORDER BY id", conpg)
    backup_file = os.path.join(backup_folder, f"tickers_pg_backup_{timestamp}.csv")
    df_tickers_pg_backup.to_csv(backup_file, index=False)
    print(f"  - Backed up tickers: {len(df_tickers_pg_backup)} rows to {backup_file}")

# Backup existing stocks if it exists
if table_exists(conpg, 'stocks'):
    df_stocks_pg_backup = pd.read_sql("SELECT * FROM stocks ORDER BY id", conpg)
    backup_file = os.path.join(backup_folder, f"stocks_pg_backup_{timestamp}.csv")
    df_stocks_pg_backup.to_csv(backup_file, index=False)
    print(f"  - Backed up stocks: {len(df_stocks_pg_backup)} rows to {backup_file}")

# 2. Copy tickers table
print("\n" + "-"*60)
print("2. COPYING TICKERS TABLE")
print("-"*60)

# Get data from SQLite
print(f"  - Reading tickers from SQLite: {len(df_tickers_lt)} rows")

# Prepare data for PostgreSQL
df_tickers_copy = df_tickers_lt.copy()

# Convert datetime columns if needed
if 'created_at' in df_tickers_copy.columns:
    df_tickers_copy['created_at'] = pd.to_datetime(df_tickers_copy['created_at'], errors='coerce')
if 'updated_at' in df_tickers_copy.columns:
    df_tickers_copy['updated_at'] = pd.to_datetime(df_tickers_copy['updated_at'], errors='coerce')

# Check if table exists and drop it with CASCADE
if table_exists(conpg, 'tickers'):
    print("  - tickers table exists in PostgreSQL")
    current_count = get_row_count(conpg, 'tickers')
    print(f"    Current rows: {current_count}")
    
    # Drop the table with CASCADE to remove dependent objects
    if drop_table_cascade(conpg, 'tickers'):
        print("    Table dropped with CASCADE, ready for fresh copy")

# Insert into PostgreSQL
try:
    df_tickers_copy.to_sql('tickers', conpg, if_exists='fail', index=False)
    print(f"  - Successfully copied {len(df_tickers_copy)} tickers to PostgreSQL")
    
    # Verify count
    new_count = get_row_count(conpg, 'tickers')
    print(f"    Verification: {new_count} rows in PostgreSQL")
    
    if new_count == len(df_tickers_copy):
        print("    ✓ Verification successful")
    else:
        print(f"    ⚠ Warning: Row count mismatch")
        
except Exception as e:
    print(f"  - Failed to copy tickers: {e}")
    conpg.rollback()
    raise

# 3. Copy stocks table
print("\n" + "-"*60)
print("3. COPYING STOCKS TABLE")
print("-"*60)

# Get data from SQLite
print(f"  - Reading stocks from SQLite: {len(df_stocks_lt)} rows")

# Prepare data for PostgreSQL
df_stocks_copy = df_stocks_lt.copy()

# Convert datetime columns if needed
if 'created_at' in df_stocks_copy.columns:
    df_stocks_copy['created_at'] = pd.to_datetime(df_stocks_copy['created_at'], errors='coerce')
if 'updated_at' in df_stocks_copy.columns:
    df_stocks_copy['updated_at'] = pd.to_datetime(df_stocks_copy['updated_at'], errors='coerce')

# Ensure numeric columns are properly typed
numeric_cols = ['price', 'max_price', 'min_price', 'pe', 'pbv', 'paid_up', 
                'market_cap', 'daily_volume', 'beta']
for col in numeric_cols:
    if col in df_stocks_copy.columns:
        df_stocks_copy[col] = pd.to_numeric(df_stocks_copy[col], errors='coerce')

# Check if table exists and drop it with CASCADE
if table_exists(conpg, 'stocks'):
    print("  - stocks table exists in PostgreSQL")
    current_count = get_row_count(conpg, 'stocks')
    print(f"    Current rows: {current_count}")
    
    # Drop the table with CASCADE
    if drop_table_cascade(conpg, 'stocks'):
        print("    Table dropped with CASCADE, ready for fresh copy")

# Insert into PostgreSQL
try:
    df_stocks_copy.to_sql('stocks', conpg, if_exists='fail', index=False)
    print(f"  - Successfully copied {len(df_stocks_copy)} stocks to PostgreSQL")
    
    # Verify count
    new_count = get_row_count(conpg, 'stocks')
    print(f"    Verification: {new_count} rows in PostgreSQL")
    
    if new_count == len(df_stocks_copy):
        print("    ✓ Verification successful")
    else:
        print(f"    ⚠ Warning: Row count mismatch")
        
except Exception as e:
    print(f"  - Failed to copy stocks: {e}")
    conpg.rollback()
    raise

# 4. Verify foreign key relationship
print("\n" + "-"*60)
print("4. VERIFYING DATA INTEGRITY")
print("-"*60)

try:
    conpg.commit()
    
    # Check if all ticker_ids in stocks exist in tickers
    sql = """
    SELECT COUNT(*) as missing_count
    FROM stocks s
    LEFT JOIN tickers t ON s.ticker_id = t.id
    WHERE t.id IS NULL
    """
    result = pd.read_sql(sql, conpg)
    missing_count = result.iloc[0, 0]
    
    if missing_count == 0:
        print(f"  ✓ Foreign key integrity maintained: All {len(df_stocks_copy)} stocks have valid ticker_id")
    else:
        print(f"  ⚠ Warning: {missing_count} stocks have ticker_id that don't exist in tickers")
        
        # Show problematic records
        sql = """
        SELECT s.id, s.name, s.ticker_id
        FROM stocks s
        LEFT JOIN tickers t ON s.ticker_id = t.id
        WHERE t.id IS NULL
        LIMIT 10
        """
        problematic = pd.read_sql(sql, conpg)
        if len(problematic) > 0:
            print("    Problematic records:")
            print(problematic)
            
except Exception as e:
    print(f"  - Error verifying integrity: {e}")
    conpg.rollback()

# 5. Summary
print("\n" + "="*80)
print("COPY OPERATION COMPLETED")
print("="*80)
print(f"\nSummary:")
print(f"  - tickers table: {get_row_count(conpg, 'tickers')} rows copied")
print(f"  - stocks table: {get_row_count(conpg, 'stocks')} rows copied")
print(f"\nBackup files saved to: {backup_folder}")
print(f"  - tickers_pg_backup_{timestamp}.csv")
print(f"  - stocks_pg_backup_{timestamp}.csv")
print("\n" + "="*80)
```

```python
# --- Fix PostgreSQL tickers table for Rails compatibility ---

print("\n" + "="*80)
print("FIX POSTGRESQL TICKERS TABLE FOR RAILS")
print("="*80)

# 1. Check current constraints
sql = """
SELECT 
    tc.constraint_name,
    tc.constraint_type,
    kcu.column_name
FROM information_schema.table_constraints tc
JOIN information_schema.key_column_usage kcu
    ON tc.constraint_name = kcu.constraint_name
WHERE tc.table_name = 'tickers'
"""
df_constraints = pd.read_sql(sql, conpg)
print("Current constraints:")
print(df_constraints)

if len(df_constraints[df_constraints['constraint_type'] == 'PRIMARY KEY']) == 0:
    print("\nNo primary key found. Adding primary key...")
    
    # Add primary key constraint
    try:
        # First, make id column NOT NULL
        conpg.execute(text("ALTER TABLE tickers ALTER COLUMN id SET NOT NULL"))
        conpg.commit()
        print("  - Set id column to NOT NULL")
        
        # Add primary key constraint
        conpg.execute(text("ALTER TABLE tickers ADD PRIMARY KEY (id)"))
        conpg.commit()
        print("  - Added PRIMARY KEY constraint")
        
        # Create sequence for auto-increment (if not exists)
        conpg.execute(text("CREATE SEQUENCE IF NOT EXISTS tickers_id_seq"))
        conpg.execute(text("ALTER TABLE tickers ALTER COLUMN id SET DEFAULT nextval('tickers_id_seq')"))
        conpg.commit()
        print("  - Set up auto-increment sequence")
        
        # Set the sequence to start from max id + 1
        result = conpg.execute(text("SELECT MAX(id) FROM tickers"))
        max_id = result.fetchone()[0]
        if max_id:
            conpg.execute(text(f"SELECT setval('tickers_id_seq', {max_id})"))
            conpg.commit()
            print(f"  - Set sequence to start from {max_id + 1}")
            
        print("\n✓ Tickers table now has proper primary key setup")
        
    except Exception as e:
        print(f"Error fixing tickers table: {e}")
        conpg.rollback()
else:
    print("\n✓ Primary key already exists")

# 2. Check stocks table
print("\n" + "-"*60)
print("Checking stocks table...")
print("-"*60)

sql = """
SELECT 
    tc.constraint_name,
    tc.constraint_type,
    kcu.column_name
FROM information_schema.table_constraints tc
JOIN information_schema.key_column_usage kcu
    ON tc.constraint_name = kcu.constraint_name
WHERE tc.table_name = 'stocks'
"""
df_constraints_stocks = pd.read_sql(sql, conpg)
print("Stocks table constraints:")
print(df_constraints_stocks)

# Add foreign key constraint if missing
if len(df_constraints_stocks[df_constraints_stocks['constraint_type'] == 'FOREIGN KEY']) == 0:
    print("\nAdding foreign key constraint to stocks table...")
    try:
        conpg.execute(text("""
            ALTER TABLE stocks 
            ADD CONSTRAINT fk_stocks_ticker_id 
            FOREIGN KEY (ticker_id) 
            REFERENCES tickers(id)
        """))
        conpg.commit()
        print("  - Added foreign key constraint")
    except Exception as e:
        print(f"Error adding foreign key: {e}")
        conpg.rollback()
else:
    print("\n✓ Foreign key already exists")

# 3. Verify the fix
print("\n" + "-"*60)
print("Verifying fix...")
print("-"*60)

# Check tickers table structure
sql = """
SELECT 
    column_name, 
    data_type, 
    is_nullable,
    column_default
FROM information_schema.columns
WHERE table_name = 'tickers'
ORDER BY ordinal_position
"""
df_verify = pd.read_sql(sql, conpg)
print("Updated tickers table structure:")
print(df_verify)

# Test Rails find by id
print("\n" + "-"*60)
print("Testing Rails compatibility...")
print("-"*60)

test_id = 747
sql = f"""
SELECT * FROM tickers WHERE id = {test_id}
"""
df_test = pd.read_sql(sql, conpg)
if len(df_test) > 0:
    print(f"✓ Ticker with id {test_id} exists: {df_test['name'].values[0]}")
    print(f"  - id is NOT NULL: {df_test['id'].values[0] is not None}")
    print(f"  - id is unique: {len(df_test)} row(s) found")
else:
    print(f"✗ Ticker with id {test_id} not found")

print("\n" + "="*80)
print("DATABASE FIX COMPLETED")
print("="*80)
print("\nNow you should be able to access:")
print("  http://localhost:5000/tickers/747")
print("\nIf you still have issues, try restarting your Rails server:")
print("  Ctrl+C to stop, then `rails server` again")
```

```python
# --- Fix PostgreSQL stocks table for Rails compatibility ---

print("\n" + "="*80)
print("FIX POSTGRESQL STOCKS TABLE FOR RAILS")
print("="*80)

# 1. Check current constraints for stocks table
sql = """
SELECT 
    tc.constraint_name,
    tc.constraint_type,
    kcu.column_name
FROM information_schema.table_constraints tc
JOIN information_schema.key_column_usage kcu
    ON tc.constraint_name = kcu.constraint_name
WHERE tc.table_name = 'stocks'
"""
df_constraints = pd.read_sql(sql, conpg)
print("Current constraints on stocks table:")
print(df_constraints)

# Check if primary key exists
pk_exists = len(df_constraints[df_constraints['constraint_type'] == 'PRIMARY KEY']) > 0

if not pk_exists:
    print("\nNo primary key found on stocks table. Adding primary key...")
    
    try:
        # First, make id column NOT NULL
        conpg.execute(text("ALTER TABLE stocks ALTER COLUMN id SET NOT NULL"))
        conpg.commit()
        print("  - Set id column to NOT NULL")
        
        # Add primary key constraint
        conpg.execute(text("ALTER TABLE stocks ADD PRIMARY KEY (id)"))
        conpg.commit()
        print("  - Added PRIMARY KEY constraint")
        
        # Create sequence for auto-increment
        conpg.execute(text("CREATE SEQUENCE IF NOT EXISTS stocks_id_seq"))
        conpg.execute(text("ALTER TABLE stocks ALTER COLUMN id SET DEFAULT nextval('stocks_id_seq')"))
        conpg.commit()
        print("  - Set up auto-increment sequence")
        
        # Set the sequence to start from max id + 1
        result = conpg.execute(text("SELECT MAX(id) FROM stocks"))
        max_id = result.fetchone()[0]
        if max_id:
            conpg.execute(text(f"SELECT setval('stocks_id_seq', {max_id})"))
            conpg.commit()
            print(f"  - Set sequence to start from {max_id + 1}")
            
        print("\n✓ Stocks table now has proper primary key setup")
        
    except Exception as e:
        print(f"Error fixing stocks table: {e}")
        conpg.rollback()
else:
    print("\n✓ Primary key already exists on stocks table")

# 2. Check foreign key relationship
print("\n" + "-"*60)
print("Checking foreign key relationship...")
print("-"*60)

# Check if foreign key constraint exists
fk_exists = len(df_constraints[df_constraints['constraint_type'] == 'FOREIGN KEY']) > 0

if not fk_exists:
    print("\nAdding foreign key constraint to stocks table...")
    try:
        conpg.execute(text("""
            ALTER TABLE stocks 
            ADD CONSTRAINT fk_stocks_ticker_id 
            FOREIGN KEY (ticker_id) 
            REFERENCES tickers(id)
        """))
        conpg.commit()
        print("  - Added foreign key constraint")
        
        # Create index on foreign key for better performance
        conpg.execute(text("CREATE INDEX IF NOT EXISTS idx_stocks_ticker_id ON stocks(ticker_id)"))
        conpg.commit()
        print("  - Created index on ticker_id")
        
    except Exception as e:
        print(f"Error adding foreign key: {e}")
        conpg.rollback()
else:
    print("\n✓ Foreign key already exists")

# 3. Verify the fix
print("\n" + "-"*60)
print("Verifying stocks table structure...")
print("-"*60)

# Check stocks table structure
sql = """
SELECT 
    column_name, 
    data_type, 
    is_nullable,
    column_default
FROM information_schema.columns
WHERE table_name = 'stocks'
ORDER BY ordinal_position
"""
df_verify = pd.read_sql(sql, conpg)
print("Updated stocks table structure:")
print(df_verify)

# 4. Test Rails compatibility for both tables
print("\n" + "-"*60)
print("Testing Rails compatibility...")
print("-"*60)

# Test tickers table
test_id = 747
sql = f"""
SELECT * FROM tickers WHERE id = {test_id}
"""
df_test_ticker = pd.read_sql(sql, conpg)
if len(df_test_ticker) > 0:
    print(f"✓ Ticker with id {test_id} exists: {df_test_ticker['name'].values[0]}")
    print(f"  - id is NOT NULL: {df_test_ticker['id'].values[0] is not None}")
else:
    print(f"✗ Ticker with id {test_id} not found")

# Test stocks table
sql = f"""
SELECT * FROM stocks WHERE ticker_id = {test_id}
"""
df_test_stock = pd.read_sql(sql, conpg)
if len(df_test_stock) > 0:
    print(f"✓ Stock with ticker_id {test_id} exists: {df_test_stock['name'].values[0]}")
    print(f"  - id is NOT NULL: {df_test_stock['id'].values[0] is not None}")
    print(f"  - ticker_id foreign key: {df_test_stock['ticker_id'].values[0]}")
else:
    print(f"✗ Stock with ticker_id {test_id} not found")

# Test foreign key integrity
print("\n" + "-"*60)
print("Testing foreign key integrity...")
print("-"*60)

sql = """
SELECT COUNT(*) as orphaned_count
FROM stocks s
LEFT JOIN tickers t ON s.ticker_id = t.id
WHERE t.id IS NULL
"""
df_orphaned = pd.read_sql(sql, conpg)
orphaned_count = df_orphaned['orphaned_count'].values[0]

if orphaned_count == 0:
    print("✓ All stocks have valid ticker_id references")
else:
    print(f"⚠ Warning: {orphaned_count} stocks have ticker_id that don't exist in tickers")

# 5. Summary
print("\n" + "="*80)
print("SUMMARY")
print("="*80)

# Check primary keys
sql = """
SELECT 
    table_name,
    constraint_type
FROM information_schema.table_constraints
WHERE table_name IN ('tickers', 'stocks')
AND constraint_type = 'PRIMARY KEY'
ORDER BY table_name
"""
df_pks = pd.read_sql(sql, conpg)
print("\nPrimary keys:")
print(df_pks)

# Check foreign keys
sql = """
SELECT 
    table_name,
    constraint_type,
    constraint_name
FROM information_schema.table_constraints
WHERE table_name = 'stocks'
AND constraint_type = 'FOREIGN KEY'
"""
df_fks = pd.read_sql(sql, conpg)
print("\nForeign keys:")
print(df_fks)

print("\n" + "="*80)
print("DATABASE FIX COMPLETED")
print("="*80)
print("\nNow you should be able to access:")
print("  http://localhost:5000/tickers/747")
print("  http://localhost:5000/stocks/747")
print("\nIf you still have issues, restart your Rails server:")
print("  Ctrl+C to stop, then `rails server` again")
```
