```python
df_name = pd.read_csv(input_file, header=None, names=['name'])
df_name.shape
```

```python
# Extend the script to list stock names in three categories

print("\n" + "="*80)
print("STOCK NAME COMPARISON ANALYSIS")
print("="*80)

# Convert to sets for easier comparison
eligibles_set = set(df_eligibles['name'].tolist())
name_set = set(df_name['name'].tolist())

# 1) Stocks that exist in both df_eligibles and df_name
both_exist = eligibles_set.intersection(name_set)
both_exist_list = sorted(list(both_exist))

print(f"\n1) Stocks existing in BOTH df_eligibles AND df_name: {len(both_exist_list)} stocks")
print("-" * 60)
if both_exist_list:
    # Print in batches for better readability
    for i in range(0, len(both_exist_list), 10):
        batch = both_exist_list[i:i+10]
        print(", ".join(batch))
else:
    print("No stocks found in both sets")

# 2) Stocks that exist only in df_eligibles (not in df_name)
only_in_eligibles = eligibles_set - name_set
only_in_eligibles_list = sorted(list(only_in_eligibles))

print(f"\n2) Stocks existing ONLY in df_eligibles: {len(only_in_eligibles_list)} stocks")
print("-" * 60)
if only_in_eligibles_list:
    for i in range(0, len(only_in_eligibles_list), 10):
        batch = only_in_eligibles_list[i:i+10]
        print(", ".join(batch))
else:
    print("No stocks found only in df_eligibles")

# 3) Stocks that exist only in df_name (not in df_eligibles)
only_in_name = name_set - eligibles_set
only_in_name_list = sorted(list(only_in_name))

print(f"\n3) Stocks existing ONLY in df_name: {len(only_in_name_list)} stocks")
print("-" * 60)
if only_in_name_list:
    for i in range(0, len(only_in_name_list), 10):
        batch = only_in_name_list[i:i+10]
        print(", ".join(batch))
else:
    print("No stocks found only in df_name")

# Summary statistics
print("\n" + "="*80)
print("SUMMARY")
print("="*80)
print(f"Total stocks in df_eligibles: {len(eligibles_set)}")
print(f"Total stocks in df_name: {len(name_set)}")
print(f"Stocks in both: {len(both_exist)}")
print(f"Stocks only in df_eligibles: {len(only_in_eligibles)}")
print(f"Stocks only in df_name: {len(only_in_name)}")

# Optional: Create DataFrames for each category if you want to save them
df_both = pd.DataFrame({'name': both_exist_list})
df_only_eligibles = pd.DataFrame({'name': only_in_eligibles_list})
df_only_name = pd.DataFrame({'name': only_in_name_list})

print("\n" + "="*80)
print("DATA FRAMES CREATED:")
print("="*80)
print("- df_both: {} stocks".format(len(df_both)))
print("- df_only_eligibles: {} stocks".format(len(df_only_eligibles)))
print("- df_only_name: {} stocks".format(len(df_only_name)))

# Optional: Save to CSV files
output_path = dat_path  # or specify another path
print(f"\nSaving CSV files to: {output_path}")

try:
    df_both.to_csv(os.path.join(output_path, 'stocks_in_both.csv'), index=False)
    print(f"  - Saved: stocks_in_both.csv ({len(df_both)} stocks)")
    
    df_only_eligibles.to_csv(os.path.join(output_path, 'stocks_only_in_eligibles.csv'), index=False)
    print(f"  - Saved: stocks_only_in_eligibles.csv ({len(df_only_eligibles)} stocks)")
    
    df_only_name.to_csv(os.path.join(output_path, 'stocks_only_in_name.csv'), index=False)
    print(f"  - Saved: stocks_only_in_name.csv ({len(df_only_name)} stocks)")
    
    print("All CSV files saved successfully!")
except Exception as e:
    print(f"Error saving CSV files: {e}")
```
