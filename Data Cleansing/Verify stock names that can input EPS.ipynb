```python
# --- Backup and Delete Delisted Stocks ---

print("\n" + "="*80)
print("BACKUP AND DELETE DELISTED STOCKS")
print("="*80)

# Ensure df_delists is available (already loaded in cell 28)
delist_names = df_delists['name'].tolist()
print(f"Delisted stocks to remove: {delist_names}")

if not delist_names:
    print("No delisted stocks to process.")
else:
    # 1. Create backups of current tables
    backup_path = dat_path  # or a dedicated backup folder
    print(f"\nCreating backups in: {backup_path}")
    
    # Backup stocks table
    stocks_backup_file = os.path.join(backup_path, 'stocks_backup_before_delete.csv')
    df_stocks.to_csv(stocks_backup_file, index=False)
    print(f"  - Backed up stocks table ({len(df_stocks)} rows) to: {stocks_backup_file}")
    
    # Backup tickers table
    tickers_backup_file = os.path.join(backup_path, 'tickers_backup_before_delete.csv')
    df_tickers.to_csv(tickers_backup_file, index=False)
    print(f"  - Backed up tickers table ({len(df_tickers)} rows) to: {tickers_backup_file}")
```

```python
# 2. Perform deletion inside a nested transaction (savepoint)
print("\nDeleting delisted stocks...")

# Use begin_nested to create a savepoint within the existing transaction
with conlt.begin_nested():
    try:
        # Delete from stocks first (child table, may have FK to tickers)
        for name in delist_names:
            delete_stocks_sql = text("DELETE FROM stocks WHERE name = :name")
            result = conlt.execute(delete_stocks_sql, {"name": name})
            print(f"    Deleted {result.rowcount} row(s) from stocks for {name}")
        
        # Delete from tickers
        for name in delist_names:
            delete_tickers_sql = text("DELETE FROM tickers WHERE name = :name")
            result = conlt.execute(delete_tickers_sql, {"name": name})
            print(f"    Deleted {result.rowcount} row(s) from tickers for {name}")
        
        # If we reach here, the savepoint will be released (i.e., changes become permanent)
        print("\nDeletion completed successfully. Changes committed to the database.")
    except Exception as e:
        # If an exception occurs, the savepoint is rolled back automatically
        print(f"\nError during deletion: {e}")
        print("Changes rolled back within the savepoint.")
        raise  # re-raise to see full error
```

```python
    # 3. Verify deletions by re-querying the tables
    print("\nVerifying deletions...")
    
    # Reload df_stocks and df_tickers from database
    df_stocks_after = pd.read_sql("SELECT * FROM stocks ORDER BY name", conlt)
    df_tickers_after = pd.read_sql("SELECT * FROM tickers ORDER BY name", conlt)
    
    print(f"Stocks table now has {len(df_stocks_after)} rows (was {len(df_stocks)}).")
    print(f"Tickers table now has {len(df_tickers_after)} rows (was {len(df_tickers)}).")
    
    # Optionally, update the in-memory dataframes for the rest of the notebook
    df_stocks = df_stocks_after
    df_tickers = df_tickers_after
    
    print("\nBackup and deletion process finished.")
```
