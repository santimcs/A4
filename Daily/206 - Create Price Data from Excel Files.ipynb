```python
### First equals to latest published pdf file
df = df.drop_duplicates(subset='Symbol', keep='first')
df.shape
```

```python
df.rename(columns={col: col.replace(" ", "-") for col in df.columns}, inplace=True)
df.columns
```

```python
cols = 'Symbol Last Change %Change Open High Low Total-Volume Total-Value-(k)'.split()
#df[cols].sort_values('Symbol',ascending=True)
df[cols].shape
```

```python
# Clean Symbol column (remove <XD>)
df['Symbol'] = df['Symbol'].str.replace(r'\s*<.*?>', '', regex=True)

# Then proceed with numeric conversion as above
numeric_cols = df.columns.drop('Symbol')
df[numeric_cols] = (
    df[numeric_cols].replace('-', 0)
    .apply(pd.to_numeric, errors='coerce', downcast=None)  # Disable downcasting
)
df[cols].shape
```
