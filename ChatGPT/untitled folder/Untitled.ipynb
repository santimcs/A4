```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import json
import pandas_datareader.data as web
from datetime import datetime

start = datetime(2010,1,1)
end = datetime(2026,4,1)

wti = web.DataReader('DCOILWTICO', 'fred', start, end)
brent = web.DataReader('DCOILBRENTEU', 'fred', start, end)

prices = pd.concat([wti, brent], axis=1)
prices.columns = ['WTI', 'Brent']
prices = prices.dropna()
prices['Spread'] = prices['Brent'] - prices['WTI']

prices.to_csv('brent_wti_spread_2010_2026.csv')

fig = px.line(prices, x=prices.index, y='Spread', title='Brent-WTI spread (2010-2026)')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Spread USD/bbl')
fig.update_traces(hovertemplate='%{y:.2f}')

fig.write_image('brent_wti_spread_2010_2026.png')
meta = {"caption": "Brent minus WTI daily spread 2010-2026", "description": "Line chart of Brent-WTI crude oil spread in USD per barrel, using FRED spot price series."}
with open('brent_wti_spread_2010_2026.png.meta.json', 'w') as f:
    json.dump(meta, f)
```
