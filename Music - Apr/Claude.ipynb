# MP4 Total Duration Calculator
Calculates the total duration of all MP4 files in a specified folder.

```python
import os
from pathlib import Path
from moviepy.editor import VideoFileClip

# ── Configuration ──────────────────────────────────────────────
FOLDER = r"C:\Users\User\OneDrive\A4\Data\videos\Easy Listening"
# ───────────────────────────────────────────────────────────────

folder = Path(FOLDER)
mp4_files = sorted(folder.glob("*.mp4"))

if not mp4_files:
    print(f"No MP4 files found in:\n  {folder}")
else:
    print(f"Found {len(mp4_files)} MP4 file(s) in:\n  {folder}\n")
    print(f"{'#':<4} {'File':<55} {'Duration':>12}")
    print("-" * 73)

    total_seconds = 0.0
    errors = []

    for i, path in enumerate(mp4_files, 1):
        try:
            with VideoFileClip(str(path)) as clip:
                dur = clip.duration          # seconds (float)
            total_seconds += dur
            h, rem = divmod(int(dur), 3600)
            m, s   = divmod(rem, 60)
            dur_str = f"{h:02d}:{m:02d}:{s:02d}" if h else f"{m:02d}:{s:02d}"
            print(f"{i:<4} {path.name:<55} {dur_str:>12}")
        except Exception as e:
            errors.append((path.name, str(e)))
            print(f"{i:<4} {path.name:<55} {'ERROR':>12}")

    # ── Summary ────────────────────────────────────────────────
    print("-" * 73)
    th, trem = divmod(int(total_seconds), 3600)
    tm, ts   = divmod(trem, 60)
    print(f"{'TOTAL DURATION':<59} {th:02d}:{tm:02d}:{ts:02d}")
    print(f"\n  = {total_seconds:.1f} seconds")
    print(f"  = {total_seconds/60:.2f} minutes")
    print(f"  = {total_seconds/3600:.4f} hours")

    if errors:
        print("\n⚠️  Files with errors:")
        for name, msg in errors:
            print(f"   • {name}: {msg}")
```
