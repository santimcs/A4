```python
import shutil
import os
from datetime import datetime

# Source and destination paths
source_file = r"c:\ruby\expense\db\development.sqlite3"
backup_base_dir = r"c:\Users\User\OneDrive\Backup\expense"

# Create a timestamped backup filename
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
destination_file_with_timestamp = os.path.join(backup_base_dir, f"development_{timestamp}.sqlite3")
destination_file_latest = os.path.join(backup_base_dir, "development.sqlite3")

# Create backup directory if it doesn't exist
os.makedirs(backup_base_dir, exist_ok=True)

try:
    # Check if source file exists
    if not os.path.exists(source_file):
        raise FileNotFoundError(f"Source file not found: {source_file}")
    
    # Get source file size before copy
    source_size = os.path.getsize(source_file)
    source_modified = datetime.fromtimestamp(os.path.getmtime(source_file))
    
    print(f"📁 Source file: {source_file}")
    print(f"   Size: {source_size:,} bytes")
    print(f"   Last modified: {source_modified}")
    print()
    
    # Copy with timestamp
    shutil.copy2(source_file, destination_file_with_timestamp)
    print(f"✅ Backup with timestamp: {destination_file_with_timestamp}")
    
    # Copy as latest version
    shutil.copy2(source_file, destination_file_latest)
    print(f"✅ Latest backup: {destination_file_latest}")
    
    # Verify both copies
    for dest_file in [destination_file_with_timestamp, destination_file_latest]:
        if os.path.exists(dest_file):
            dest_size = os.path.getsize(dest_file)
            if source_size == dest_size:
                print(f"   ✅ Verified: {os.path.basename(dest_file)} ({dest_size:,} bytes)")
            else:
                print(f"   ⚠️ Size mismatch: {os.path.basename(dest_file)}")
    
    print("\n🎉 Backup completed successfully!")
    
except FileNotFoundError as e:
    print(f"❌ Error: {e}")
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import shutil
import os
import hashlib

def copy_sqlite_backup():
    source_file = r"c:\ruby\expense\db\development.sqlite3"
    destination_dir = r"c:\Users\User\OneDrive\Backup\expense"
    destination_file = os.path.join(destination_dir, "development.sqlite3")
    
    # Create destination directory
    os.makedirs(destination_dir, exist_ok=True)
    
    # Check if source exists
    if not os.path.exists(source_file):
        print(f"❌ Source file not found: {source_file}")
        return False
    
    # Check if source is a SQLite file (first 16 bytes should be "SQLite format 3")
    try:
        with open(source_file, 'rb') as f:
            header = f.read(16)
            if header != b'SQLite format 3\x00':
                print("⚠️ Warning: Source file may not be a valid SQLite database")
    except Exception as e:
        print(f"⚠️ Warning: Could not verify SQLite header: {e}")
    
    try:
        # Calculate MD5 hash of source file
        md5_hash = hashlib.md5()
        with open(source_file, 'rb') as f:
            for chunk in iter(lambda: f.read(4096), b""):
                md5_hash.update(chunk)
        source_md5 = md5_hash.hexdigest()
        
        # Copy the file
        shutil.copy2(source_file, destination_file)
        print(f"✅ File copied: {destination_file}")
        
        # Verify with MD5 hash
        md5_hash = hashlib.md5()
        with open(destination_file, 'rb') as f:
            for chunk in iter(lambda: f.read(4096), b""):
                md5_hash.update(chunk)
        dest_md5 = md5_hash.hexdigest()
        
        if source_md5 == dest_md5:
            print(f"✅ MD5 checksum verified: {source_md5}")
            print(f"🎉 Backup successful!")
            return True
        else:
            print(f"❌ MD5 mismatch!")
            print(f"   Source: {source_md5}")
            print(f"   Destination: {dest_md5}")
            return False
            
    except PermissionError:
        print(f"❌ Permission denied. Is the SQLite database in use?")
        return False
    except Exception as e:
        print(f"❌ Error during copy: {e}")
        return False

# Run the backup
copy_sqlite_backup()
```
