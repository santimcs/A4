```python
import os
import datetime
import pandas as pd
from moviepy.editor import VideoFileClip

def calculate_total_duration(directory, export_csv=True):
"""
Scans a directory for .mp4 files, calculates total duration,
and exports the results to a CSV file.
"""
total_seconds = 0
video_data = []
# Verify if directory exists
if not os.path.exists(directory):
    print(f"Error: The directory '{directory}' does not exist.")
    return

print(f"Scanning directory: {directory}\n")

# List all files in the directory
for filename in os.listdir(directory):
    if filename.lower().endswith(".mp4"):
        filepath = os.path.join(directory, filename)
        try:
            # Load the video clip to get duration
            with VideoFileClip(filepath) as clip:
                duration = clip.duration
                total_seconds += duration
                
                # Collect data for the report
                video_data.append({
                    "Filename": filename,
                    "Duration (Seconds)": round(duration, 2),
                    "Duration (Formatted)": str(datetime.timedelta(seconds=int(duration)))
                })
                print(f"Processed: {filename}")
        except Exception as e:
            print(f"Could not process {filename}: {e}")

# Create a DataFrame for easy viewing and export
df = pd.DataFrame(video_data)

# Calculate the final formatted total time
total_time_str = str(datetime.timedelta(seconds=int(total_seconds)))

print("\n" + "="*40)
print(f"Total MP4 files found: {len(video_data)}")
print(f"Total Duration: {total_time_str} (HH:MM:SS)")
print("="*40)

# Save to CSV if requested
if export_csv and not df.empty:
    output_path = os.path.join(directory, "video_durations_report.csv")
    df.to_csv(output_path, index=False)
    print(f"\nSuccess! Report saved to: {output_path}")

return df
```
