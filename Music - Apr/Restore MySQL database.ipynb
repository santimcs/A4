```python
import subprocess
import os

# Configuration
backup_dir = r"C:\Users\PC1\OneDrive\Documents\Backup\mysql"
sql_file = "music_development_20260403.sql"
database = "music_development"
username = "root"
password = ""  # Blank password

# Full path to SQL file
sql_file_path = os.path.join(backup_dir, sql_file)

# MySQL executable path (using forward slashes or raw string)
mysql_exe = r"C:\xampp\MySQL\bin\mysql.exe"

# Check if SQL file exists
if not os.path.exists(sql_file_path):
    raise FileNotFoundError(f"SQL file not found: {sql_file_path}")

print(f"Importing {sql_file_path} into database '{database}'...")

# Build command
# For blank password, omit -p flag entirely (or use -p with no value)
cmd = [mysql_exe, f"-u{username}", database]

# Execute with input redirection from SQL file
try:
    with open(sql_file_path, 'r', encoding='utf-8') as f:
        result = subprocess.run(
            cmd,
            stdin=f,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
    
    if result.returncode == 0:
        print("✅ Import completed successfully!")
    else:
        print(f"❌ Import failed with exit code {result.returncode}")
        print(f"Error output:\n{result.stderr}")
        
except Exception as e:
    print(f"❌ Error: {e}")
```
