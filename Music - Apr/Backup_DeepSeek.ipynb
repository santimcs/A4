```python
import os
import subprocess
from datetime import datetime

# Hardcoded values
database = "music_development"  # CHANGE THIS to your database name
backup_file = rf"C:\Users\User\OneDrive\Documents\Backup\mysql\{database}_{datetime.now().strftime('%Y%m%d')}.sql"

# Create directory if it doesn't exist
os.makedirs(r"C:\Users\User\OneDrive\Documents\Backup\mysql", exist_ok=True)

# Run mysqldump with no password
cmd = r'c:\xampp\mysql\bin\mysqldump -uroot ' + database

print(f"Running: {cmd}")
print(f"Output to: {backup_file}")

# Execute and save output
with open(backup_file, 'w') as f:
    result = subprocess.run(cmd, shell=True, stdout=f, stderr=subprocess.PIPE, text=True)

if result.returncode == 0:
    print(f"✅ Success! Backup saved to {backup_file}")
    print(f"File size: {os.path.getsize(backup_file):,} bytes")
else:
    print(f"❌ Failed: {result.stderr}")
```
