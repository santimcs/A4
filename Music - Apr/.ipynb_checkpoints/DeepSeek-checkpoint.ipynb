```python
# Cell 1: Import required libraries
import os
import subprocess
import json
from pathlib import Path

# Cell 2: Define the folder path and function to get video duration
folder_path = r"C:\Users\User\OneDrive\A4\Data\videos\Easy Listening"

def get_mp4_duration_ffprobe(filepath):
    """Get duration of MP4 file using ffprobe"""
    try:
        cmd = [
            'ffprobe',
            '-v', 'error',
            '-show_entries', 'format=duration',
            '-of', 'json',
            filepath
        ]
        result = subprocess.run(cmd, capture_output=True, text=True)
        data = json.loads(result.stdout)
        return float(data['format']['duration'])
    except:
        return None

def get_mp4_duration_alternative(filepath):
    """Alternative method using moviepy if ffprobe not available"""
    try:
        from moviepy.editor import VideoFileClip
        with VideoFileClip(filepath) as video:
            return video.duration
    except:
        return None

# Cell 3: Find all MP4 files in the folder
mp4_files = list(Path(folder_path).glob("*.mp4"))
mp4_files.extend(Path(folder_path).glob("*.MP4"))  # Also check uppercase extension

print(f"Found {len(mp4_files)} MP4 files in: {folder_path}\n")

# Cell 4: Calculate durations for each file
video_durations = {}
total_duration_seconds = 0

for i, video_file in enumerate(mp4_files, 1):
    print(f"Processing {i}/{len(mp4_files)}: {video_file.name}")
    
    # Try ffprobe first
    duration = get_mp4_duration_ffprobe(str(video_file))
    
    # If ffprobe fails, try alternative method
    if duration is None:
        duration = get_mp4_duration_alternative(str(video_file))
    
    if duration is not None:
        video_durations[video_file.name] = duration
        total_duration_seconds += duration
    else:
        print(f"  Warning: Could not get duration for {video_file.name}")

# Cell 5: Format and display results
def format_time(seconds):
    """Convert seconds to HH:MM:SS format"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{secs:02d}"

print("=" * 60)
print("VIDEO DURATION SUMMARY")
print("=" * 60)
print(f"\nTotal number of MP4 files: {len(video_durations)}")
print(f"Total duration (seconds): {total_duration_seconds:.2f}")
print(f"Total duration (HH:MM:SS): {format_time(total_duration_seconds)}")
print(f"Total duration (minutes): {total_duration_seconds / 60:.2f}")
print(f"Total duration (hours): {total_duration_seconds / 3600:.2f}")

# Cell 6: Display individual file durations (fixed Unicode handling)
print("\n" + "=" * 60)
print("INDIVIDUAL FILE DURATIONS")
print("=" * 60)
print(f"{'File Name':<50} {'Duration (HH:MM:SS)':<20}")
print("-" * 70)

for filename, duration in video_durations.items():
    # Truncate filename if too long and handle Unicode
    display_name = filename if len(filename) <= 47 else filename[:44] + "..."
    # Ensure the string can be printed (encode/decode to handle Unicode)
    try:
        print(f"{display_name:<50} {format_time(duration):<20}")
    except UnicodeEncodeError:
        # Fallback: replace problematic characters
        safe_name = display_name.encode('ascii', 'ignore').decode('ascii')
        print(f"{safe_name:<50} {format_time(duration):<20}")

# Cell 7: Save results to a text file with UTF-8 encoding (fixed)
output_file = os.path.join(folder_path, "video_durations_summary.txt")

# Use UTF-8 encoding to handle all Unicode characters
with open(output_file, 'w', encoding='utf-8') as f:
    f.write("VIDEO DURATION SUMMARY\n")
    f.write("=" * 60 + "\n")
    f.write(f"Folder: {folder_path}\n")
    f.write(f"Total MP4 files: {len(video_durations)}\n")
    f.write(f"Total duration (seconds): {total_duration_seconds:.2f}\n")
    f.write(f"Total duration (HH:MM:SS): {format_time(total_duration_seconds)}\n")
    f.write(f"Total duration (minutes): {total_duration_seconds / 60:.2f}\n")
    f.write(f"Total duration (hours): {total_duration_seconds / 3600:.2f}\n\n")
    f.write("INDIVIDUAL FILE DURATIONS\n")
    f.write("-" * 60 + "\n")
    for filename, duration in video_durations.items():
        f.write(f"{filename:<50} {format_time(duration)}\n")

print(f"\nResults saved to: {output_file}")

# Cell 8: Alternative - Create a simple CSV report (also with UTF-8)
csv_file = os.path.join(folder_path, "video_durations_report.csv")

with open(csv_file, 'w', encoding='utf-8-sig') as f:  # utf-8-sig for Excel compatibility
    f.write("Filename,Duration (seconds),Duration (HH:MM:SS)\n")
    for filename, duration in video_durations.items():
        f.write(f'"{filename}",{duration:.2f},{format_time(duration)}\n')

print(f"CSV report saved to: {csv_file}")

# Cell 9: Display summary statistics
print("\n" + "=" * 60)
print("STATISTICS")
print("=" * 60)
if len(video_durations) > 0:
    durations_list = list(video_durations.values())
    print(f"Average duration: {format_time(sum(durations_list) / len(durations_list))}")
    print(f"Shortest video: {format_time(min(durations_list))}")
    print(f"Longest video: {format_time(max(durations_list))}")
```

```python
# Cell 1: Import required libraries
import os
import subprocess
import json
from pathlib import Path
from collections import defaultdict

# Cell 2: Define the folder path and functions
folder_path = r"C:\Users\User\OneDrive\A4\Data\videos\Easy Listening"

def get_mp4_duration_ffprobe(filepath):
    """Get duration of MP4 file using ffprobe"""
    try:
        cmd = [
            'ffprobe',
            '-v', 'error',
            '-show_entries', 'format=duration',
            '-of', 'json',
            filepath
        ]
        result = subprocess.run(cmd, capture_output=True, text=True)
        data = json.loads(result.stdout)
        return float(data['format']['duration'])
    except:
        return None

def normalize_filename(filename):
    """Remove ' - Copy' and other duplicate indicators from filename"""
    # Remove ' - Copy' pattern
    name = filename.replace(' - Copy', '')
    name = name.replace('(Copy)', '')
    name = name.replace(' copy', '')
    # Strip extra spaces
    name = ' '.join(name.split())
    return name

# Cell 3: Find all MP4 files and identify duplicates
mp4_files = list(Path(folder_path).glob("*.mp4"))
mp4_files.extend(Path(folder_path).glob("*.MP4"))

print(f"Total files found (including duplicates): {len(mp4_files)}")
print(f"Folder path: {folder_path}\n")

# Group files by normalized name
file_groups = defaultdict(list)
for video_file in mp4_files:
    normalized_name = normalize_filename(video_file.stem) + video_file.suffix
    file_groups[normalized_name].append(video_file)

print(f"Unique videos (after removing duplicates): {len(file_groups)}")
print("\nDuplicate detection:")
for normalized_name, files in file_groups.items():
    if len(files) > 1:
        print(f"\n  '{normalized_name}' appears {len(files)} times:")
        for f in files:
            print(f"    - {f.name}")

# Cell 4: Calculate durations for unique videos only
video_durations = {}
total_duration_seconds = 0
duplicates_skipped = 0

print("\n" + "=" * 60)
print("PROCESSING UNIQUE VIDEOS")
print("=" * 60)

for normalized_name, files in file_groups.items():
    # Use the first file (prefer non-copy version if available)
    # Prioritize files without 'Copy' in name
    primary_file = None
    for f in files:
        if 'Copy' not in f.name:
            primary_file = f
            break
    if primary_file is None:
        primary_file = files[0]  # Use first if all are copies
    
    if len(files) > 1:
        duplicates_skipped += len(files) - 1
        print(f"\nProcessing: {normalized_name}")
        print(f"  Using: {primary_file.name}")
        print(f"  Skipping {len(files)-1} duplicate(s)")
    else:
        print(f"\nProcessing: {primary_file.name}")
    
    # Get duration
    duration = get_mp4_duration_ffprobe(str(primary_file))
    
    if duration is not None:
        video_durations[normalized_name] = {
            'duration': duration,
            'original_filename': primary_file.name,
            'file_path': str(primary_file)
        }
        total_duration_seconds += duration
    else:
        print(f"  Warning: Could not get duration for {primary_file.name}")

# Cell 5: Format and display results
def format_time(seconds):
    """Convert seconds to HH:MM:SS format"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{secs:02d}"

print("\n" + "=" * 60)
print("VIDEO DURATION SUMMARY")
print("=" * 60)
print(f"\nTotal files scanned: {len(mp4_files)}")
print(f"Duplicate files skipped: {duplicates_skipped}")
print(f"Unique videos processed: {len(video_durations)}")
print(f"\nTotal duration (seconds): {total_duration_seconds:.2f}")
print(f"Total duration (HH:MM:SS): {format_time(total_duration_seconds)}")
print(f"Total duration (minutes): {total_duration_seconds / 60:.2f}")
print(f"Total duration (hours): {total_duration_seconds / 3600:.2f}")

# Cell 6: Display individual file durations
print("\n" + "=" * 60)
print("INDIVIDUAL FILE DURATIONS (UNIQUE VIDEOS)")
print("=" * 60)
print(f"{'#':<3} {'Video Name':<45} {'Duration':<12}")
print("-" * 60)

for i, (name, info) in enumerate(video_durations.items(), 1):
    # Truncate name if too long
    display_name = name if len(name) <= 42 else name[:39] + "..."
    print(f"{i:<3} {display_name:<45} {format_time(info['duration']):<12}")

# Cell 7: Save detailed report
output_file = os.path.join(folder_path, "video_durations_unique_summary.txt")

with open(output_file, 'w', encoding='utf-8') as f:
    f.write("VIDEO DURATION SUMMARY (DUPLICATES REMOVED)\n")
    f.write("=" * 60 + "\n")
    f.write(f"Folder: {folder_path}\n")
    f.write(f"Total files scanned: {len(mp4_files)}\n")
    f.write(f"Duplicate files skipped: {duplicates_skipped}\n")
    f.write(f"Unique videos processed: {len(video_durations)}\n")
    f.write(f"\nTotal duration (seconds): {total_duration_seconds:.2f}\n")
    f.write(f"Total duration (HH:MM:SS): {format_time(total_duration_seconds)}\n")
    f.write(f"Total duration (minutes): {total_duration_seconds / 60:.2f}\n")
    f.write(f"Total duration (hours): {total_duration_seconds / 3600:.2f}\n\n")
    f.write("INDIVIDUAL FILE DETAILS\n")
    f.write("-" * 60 + "\n")
    for name, info in video_durations.items():
        f.write(f"\nVideo: {name}\n")
        f.write(f"  Filename: {info['original_filename']}\n")
        f.write(f"  Duration: {format_time(info['duration'])} ({info['duration']:.2f} seconds)\n")

print(f"\n\nDetailed report saved to: {output_file}")

# Cell 8: Find and list duplicate files for cleanup
print("\n" + "=" * 60)
print("DUPLICATE FILES FOR REVIEW")
print("=" * 60)

duplicate_files = []
for normalized_name, files in file_groups.items():
    if len(files) > 1:
        print(f"\nOriginal/Clean: {normalized_name}")
        for f in files:
            size_mb = os.path.getsize(f) / (1024 * 1024)
            is_copy = "COPY" in f.name.upper()
            marker = " ⚠️ DUPLICATE" if is_copy else " ✓ PRIMARY"
            print(f"  {marker}: {f.name} ({size_mb:.2f} MB)")
            if is_copy:
                duplicate_files.append(f)

if duplicate_files:
    print(f"\n\nFound {len(duplicate_files)} duplicate file(s) that can potentially be deleted:")
    for f in duplicate_files:
        print(f"  - {f.name}")
    print("\n⚠️  WARNING: Review these files before deleting!")
    print("   They may be OneDrive sync artifacts or intentional copies.")
else:
    print("\nNo duplicate files found.")
```
