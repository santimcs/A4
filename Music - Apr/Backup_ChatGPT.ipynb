```python
import os
import sys
import subprocess
from datetime import datetime
```

```python
# Simulate ARGV
database = "music_development"   # e.g. "mydb"
end_of_iter = 'Apr_3'               # or e.g. "iteration1"

username = "root"
backup_dir = r"C:\Users\User\OneDrive\Documents\Backup\mysql"
```

```python
try:
    # Ensure backup directory exists
    if not os.path.exists(backup_dir):
        os.makedirs(backup_dir)

    # Create backup filename
    if end_of_iter is None:
        backup_file = os.path.join(
            backup_dir,
            f"{database}_{datetime.now().strftime('%Y%m%d')}"
        )
    else:
        backup_file = os.path.join(
            backup_dir,
            f"{database}_{end_of_iter}"
        )

    output_file = f"{backup_file}.sql"

    # Build mysqldump command
    password = ""

    cmd = [
        r"c:\xampp\mysql\bin\mysqldump",
        f"-u{username}",
        # f"-p{password}",   # ⚠️ no space!
        database
    ]

    # Execute command and redirect output
    with open(output_file, "w") as f:
        result = subprocess.run(cmd, stdout=f)

    if result.returncode == 0:
        print(f"Backup completed successfully: {output_file}")
    else:
        raise Exception(f"mysqldump failed with exit code {result.returncode}")

except Exception as e:
    print(f"Error: {e}")
```
