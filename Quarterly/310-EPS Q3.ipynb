```python
from sqlalchemy import create_engine, text
from sqlalchemy.exc import OperationalError
import time

def delete_with_retry(stock_list_str, quarter, max_retries=5):
    """
    Attempt delete operation with retry logic for database locks.
    """
    # Create engine with longer timeout and WAL mode enabled
    engine = create_engine(
        "sqlite:///c:\\ruby\\portmy\\db\\development.sqlite3",
        connect_args={"timeout": 30}  # Wait up to 30 seconds for lock to clear
    )
    
    for attempt in range(max_retries):
        try:
            with engine.connect() as conn:
                # Enable WAL mode for better concurrency (doesn't require transaction)
                conn.execute(text("PRAGMA journal_mode=WAL"))
                
                # Execute delete in its own transaction block
                with conn.begin():  # This starts a new transaction
                    sqlDel = f"""
                        DELETE FROM profits 
                        WHERE name IN ('{stock_list_str}')
                        AND quarter < {quarter}
                    """
                    result = conn.execute(text(sqlDel))
                    rowcount = result.rowcount
                
                # Transaction auto-commits here if no errors
                print(f"✅ Success: Deleted {rowcount} rows")
                return rowcount
                
        except OperationalError as e:
            if "database is locked" in str(e) and attempt < max_retries - 1:
                print(f"⏳ Database locked, retrying in {2 ** attempt} seconds... (Attempt {attempt + 1}/{max_retries})")
                time.sleep(2 ** attempt)  # Exponential backoff
            else:
                print(f"❌ Failed after {max_retries} attempts: {e}")
                raise
        finally:
            # Ensure engine resources are cleaned up
            engine.dispose()

# Use the function
try:
    rows_deleted = delete_with_retry(stock_list_str, quarter)
except Exception as e:
    print(f"Delete operation failed: {e}")
```
