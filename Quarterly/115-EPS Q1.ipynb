```python
sql = f"""
SELECT *
FROM epss
WHERE name IN ('{stock_list_str}')
AND publish_date >= '2026-04-01'
ORDER BY name, year DESC, quarter DESC"""
df_tmp = pd.read_sql(sql, conlt)
df_tmp.shape
```

```python
df_tmp["yoy_gain"] = df_tmp["q_amt"] - df_tmp["y_amt"]
df_tmp["yoy_pct"]  = round(df_tmp["yoy_gain"] / abs(df_tmp["y_amt"]) * 100,2)
df_tmp["acc_gain"] = df_tmp["aq_amt"] - df_tmp["ay_amt"]
df_tmp["acc_pct"] = round(df_tmp["acc_gain"] / abs(df_tmp["ay_amt"]) * 100,2)
df_aggr = df_tmp[
    [
        "name",
        "year",
        "quarter",
        "q_amt",
        "y_amt",
        "yoy_gain",
        "yoy_pct",
        "aq_amt",
        "ay_amt",
        "acc_gain",
        "acc_pct",
    ]
]
```

```python
# Get unique rows based on 'name' column
result = df_tmp.drop_duplicates(subset=['name'])
result.sort_values(by=["name"], ascending=[True]).style.format(format_dict)
```
