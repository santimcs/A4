```python
# ============================================
# BACKUP AND DELETE OPERATIONS FOR epss TABLE
# ============================================

# Step 1: Create backup of epss table before deletion
print("Creating backup of epss table...")

# Method 1: Create a backup table with timestamp
from datetime import datetime
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
backup_table_name = f"epss_backup_{timestamp}"

# Create backup table with all data from epss
backup_sql = f"""
CREATE TABLE IF NOT EXISTS {backup_table_name} AS
SELECT * FROM epss
"""
conlt.execute(text(backup_sql))
conlt.commit()
print(f"Backup table '{backup_table_name}' created successfully")

# Method 2: Also export to CSV as additional backup (optional)
backup_csv_path = f"../data/epss_backup_{timestamp}.csv"
df_epss_backup = pd.read_sql("SELECT * FROM epss", conlt)
df_epss_backup.to_csv(backup_csv_path, index=False)
print(f"CSV backup saved to: {backup_csv_path}")

# Step 2: Display count of records to be deleted
if len(df_to_delete) > 0:
    stock_list_to_delete = df_to_delete['name'].tolist()
    stock_list_str_delete = "','".join(stock_list_to_delete)
    
    # Count records to be deleted
    count_sql = f"""
    SELECT COUNT(*) as count_to_delete
    FROM epss
    WHERE name IN ('{stock_list_str_delete}')
    """
    count_to_delete = pd.read_sql(count_sql, conlt).iloc[0]['count_to_delete']
    print(f"\nRecords to be deleted: {count_to_delete}")
    print(f"Stocks to be deleted: {stock_list_to_delete}")
    
    # Confirm deletion (uncomment if you want interactive confirmation)
    # confirmation = input("Do you want to proceed with deletion? (yes/no): ")
    # if confirmation.lower() != 'yes':
    #     print("Deletion cancelled")
    #     exit()
    
    # Step 3: Perform deletion
    delete_sql = f"""
    DELETE FROM epss
    WHERE name IN ('{stock_list_str_delete}')
    """
    
    try:
        result = conlt.execute(text(delete_sql))
        conlt.commit()
        print(f"\n✓ Successfully deleted {result.rowcount} records from epss table")
    except Exception as e:
        print(f"Error during deletion: {e}")
        conlt.rollback()
else:
    print("No records to delete - df_to_delete is empty")

# Step 4: Verify deletion
print("\n" + "="*50)
print("VERIFICATION AFTER DELETION")
print("="*50)

# Check remaining records
remaining_count = pd.read_sql("SELECT COUNT(*) as remaining FROM epss", conlt)
print(f"Remaining records in epss: {remaining_count.iloc[0]['remaining']}")

# Verify specific stocks were deleted
if len(df_to_delete) > 0:
    sample_stock = df_to_delete['name'].iloc[0]
    check_sql = f"SELECT COUNT(*) as found FROM epss WHERE name = '{sample_stock}'"
    check_result = pd.read_sql(check_sql, conlt)
    if check_result.iloc[0]['found'] == 0:
        print(f"✓ Verification passed: Stock '{sample_stock}' successfully deleted")
    else:
        print(f"✗ Verification failed: Stock '{sample_stock}' still exists")
```
