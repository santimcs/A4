```python
import sys
import pkg_resources

# List of packages to check (from your imports)
required_packages = [
    'moviepy',
    'sqlalchemy', 
    'pandas',
    'numpy',
    'Pillow',  # Note: PIL is actually Pillow in PyPI
    'matplotlib',
    'pymysql'  # For MySQL connection
]

print("Python Version:", sys.version)
print("\n" + "="*50)
print("Installed Package Versions:")
print("="*50)

for package in required_packages:
    try:
        version = pkg_resources.get_distribution(package).version
        print(f"{package:<15} : {version}")
    except pkg_resources.DistributionNotFound:
        print(f"{package:<15} : NOT INSTALLED")
    except Exception as e:
        print(f"{package:<15} : ERROR - {e}")

# Additional check for system info
print("\n" + "="*50)
print("System Information:")
print("="*50)
import platform
print(f"System: {platform.system()}")
print(f"Release: {platform.release()}")
print(f"Architecture: {platform.architecture()}")
print(f"Machine: {platform.machine()}")
```

```python
import sys
import os
import subprocess

print("=" * 60)
print("DIAGNOSTICS FOR MOVIEPY ISSUE")
print("=" * 60)

# 1. Check Python executable
print(f"1. Python executable: {sys.executable}")
print(f"   Python version: {sys.version}")

# 2. Check PATH
print(f"\n2. PATH environment variable:")
for i, path in enumerate(os.environ['PATH'].split(';'), 1):
    if 'python' in path.lower() or 'anaconda' in path.lower() or 'conda' in path.lower():
        print(f"   {i:3d}. {path}")

# 3. Check where moviepy is installed
print("\n3. Checking for moviepy installation...")
try:
    import moviepy
    print(f"   ✓ moviepy imported: {moviepy.__file__}")
except ImportError:
    print("   ✗ moviepy not found")
    
# 4. List all Python installations
print("\n4. All Python installations found:")
result = subprocess.run(['where', 'python'], capture_output=True, text=True)
print(result.stdout)

# 5. Check pip list
print("\n5. Installed packages in current environment:")
result = subprocess.run([sys.executable, '-m', 'pip', 'list'], capture_output=True, text=True)
# Filter for relevant packages
for line in result.stdout.split('\n'):
    if any(pkg in line.lower() for pkg in ['moviepy', 'sqlalchemy', 'pandas', 'numpy', 'pillow', 'matplotlib', 'pymysql']):
        print(f"   {line}")

# 6. Check sys.path
print("\n6. Python module search paths (first 10):")
for i, path in enumerate(sys.path[:10], 1):
    print(f"   {i:2d}. {path}")

print("\n" + "=" * 60)
```

```python
import os
import sys

moviepy_path = r"C:\Users\User\anaconda3\Lib\site-packages\moviepy"

print("Checking moviepy directory structure...")
print("=" * 60)

if os.path.exists(moviepy_path):
    print(f"✓ MoviePy found at: {moviepy_path}")
    
    # List contents
    for item in os.listdir(moviepy_path):
        item_path = os.path.join(moviepy_path, item)
        if os.path.isfile(item_path):
            print(f"  📄 {item}")
        else:
            print(f"  📁 {item}/")
    
    # Check for editor.py specifically
    editor_path = os.path.join(moviepy_path, "editor.py")
    print(f"\nLooking for editor.py...")
    if os.path.exists(editor_path):
        print(f"✓ editor.py found at: {editor_path}")
    else:
        print(f"✗ editor.py NOT found!")
        
    # Check for editor directory
    editor_dir = os.path.join(moviepy_path, "editor")
    if os.path.exists(editor_dir) and os.path.isdir(editor_dir):
        print(f"✓ editor directory found at: {editor_dir}")
        print("  Contents of editor directory:")
        for item in os.listdir(editor_dir):
            print(f"    - {item}")
else:
    print(f"✗ MoviePy directory not found!")

print("\n" + "=" * 60)

# Try to import moviepy and check its __file__
try:
    import moviepy
    print(f"✓ moviepy.__file__: {moviepy.__file__}")
    print(f"✓ moviepy.__version__: {moviepy.__version__}")
    
    # Try to see what's in moviepy
    print("\n✓ Contents of moviepy module:")
    for attr in dir(moviepy):
        if not attr.startswith('_'):
            print(f"  - {attr}")
except Exception as e:
    print(f"✗ Error: {e}")
```
