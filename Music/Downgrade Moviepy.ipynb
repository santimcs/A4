```python
import subprocess
import sys

print("Downgrading MoviePy to 1.0.3 to match New PC...")
print("This will fix ALL compatibility issues at once.")
print("=" * 60)

# Uninstall current version
result = subprocess.run([sys.executable, "-m", "pip", "uninstall", "moviepy", "-y"], 
                       capture_output=True, text=True)
print("Uninstalled current version")

# Install version 1.0.3 (same as New PC)
result = subprocess.run([sys.executable, "-m", "pip", "install", "moviepy==1.0.3"], 
                       capture_output=True, text=True)
print(result.stdout)

if result.returncode == 0:
    print("\n✓ Successfully installed MoviePy 1.0.3")
    print("🎉 Your script should now work without any modifications!")
    print("\n⚠️  Please RESTART YOUR KERNEL (Kernel → Restart)")
else:
    print("\n✗ Installation failed. Using compatibility mode...")
```

```python
import subprocess
import sys

print("=" * 60)
print("FIXING NUMPY COMPATIBILITY ISSUE")
print("=" * 60)
print("Downgrading NumPy from 2.4.0 to 1.26.4 (stable version)...")

# Uninstall current numpy
result = subprocess.run([sys.executable, "-m", "pip", "uninstall", "numpy", "-y"], 
                       capture_output=True, text=True)
print("✓ Uninstalled NumPy 2.4.0")

# Install NumPy 1.26.4 (stable, widely compatible)
result = subprocess.run([sys.executable, "-m", "pip", "install", "numpy==1.26.4"], 
                       capture_output=True, text=True)
print(result.stdout)

if result.returncode == 0:
    print("\n✓ Successfully installed NumPy 1.26.4")
    print("\n⚠️  CRITICAL: You MUST restart your kernel now!")
    print("   Go to: Kernel → Restart Kernel")
    print("\nAfter restarting, run your imports again.")
else:
    print("\n✗ Installation failed. Trying conda...")
    # Try conda if pip fails
    result = subprocess.run(["conda", "install", "numpy=1.26.4", "-y"], 
                           capture_output=True, text=True)
    print(result.stdout)
```
