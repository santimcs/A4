```python
def update_buy(name, period, grade):
    # Use parameterized query to avoid SQL injection
    sql = text("""
        UPDATE buy
        SET period = :period, 
            grade = :grade 
        WHERE name = :name
    """)
    
    # Execute the query with parameters
    rp = const.execute(sql, {
        'period': period,
        'grade': grade,
        'name': name
    })
    
    return f"Records updated = {rp.rowcount}"
```
